##**Problem 5** Python functionalization

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.

    Parameters
    ----------------
    function : function
        The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
        Specify the range in the same way as the argument of np.arange ().

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
        A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
              Function gradient. One factor is reduced to make a difference
    """
    x_array = np.arange(*x_range)
    y_array = function(x_array)

    xy_array = np.concatenate((x_array[:, np.newaxis], y_array[:, np.newaxis]), axis=1)
    gradient = (xy_array[1:, 1] - xy_array[:-1, 1]) / (xy_array[1:, 0] - xy_array[:-1, 0])

    return xy_array, gradient

def function1(x_array):
    return x_array ** 2

def function2(x_array):
    return 2 * x_array ** 2 + 2 ** x_array

def function3(x_array):
    return np.sin(x_array ** 0.5)

# Compute gradients and arrays for each function
xy_array1, gradient1 = compute_gradient(function1)
xy_array2, gradient2 = compute_gradient(function2)
xy_array3, gradient3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))
```

```python
print(compute_gradient(function1))
```

```python
# Plot results for function 1
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y=x**2")
plt.plot(xy_array1[:, 0], xy_array1[:, 1])
plt.show()

plt.xlabel("X")
plt.ylabel("gradient")
plt.title("Gradient of y=x**2")
plt.plot(xy_array1[:-1, 0], gradient1)
plt.show()
```

```python
print(compute_gradient(function2))
```

```python
# Plot results for function 2
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y=2*x**2 + 2**x")
plt.plot(xy_array2[:, 0], xy_array2[:, 1])
plt.show()

plt.xlabel("X")
plt.ylabel("gradient")
plt.title("Gradient of y=2*x**2 + 2**x")
plt.plot(xy_array2[:-1, 0], gradient2)
plt.show()
```

```python
print(compute_gradient(function3))
```

```python
# Plot results for function 3
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y=sin(x**0.5)")
plt.plot(xy_array3[:, 0], xy_array3[:, 1])
plt.show()

plt.xlabel("X")
plt.ylabel("gradient")
plt.title("Gradient of y=sin(x**0.5)")
plt.plot(xy_array3[:-1, 0], gradient3)
plt.show()
```

```python
# Find the minimum value of y and its index for each function
min_y1, min_index1 = xy_array1.min(axis=0)[1], xy_array1.argmin(axis=0)[1]
min_y2, min_index2 = xy_array2.min(axis=0)[1], xy_array2.argmin(axis=0)[1]
min_y3, min_index3 = xy_array3.min(axis=0)[1], xy_array3.argmin(axis=0)[1]

print("Minimum y value and its index for function 1:", min_y1, min_index1)
print("Minimum y value and its index for function 2:", min_y2, min_index2)
print("Minimum y value and its index for function 3:", min_y3, min_index3)


# Display the slope before and after taking the minimum y for each function
print("Slope before and after minimum y for function 1:")
print(gradient1[min_index1 - 1 : min_index1 + 2])
print("Slope before and after minimum y for function 2:")
print(gradient2[min_index2 - 1 : min_index2 + 2])
print("Slope before and after minimum y for function 3:")
print(gradient3[min_index3 - 1 : min_index3 + 2])
```
