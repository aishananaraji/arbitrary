```python
import math
import matplotlib.pyplot as plt

CHESTNUT_DIAMETER_CM = 2.54
CHESTNUT_RADIUS_CM = CHESTNUT_DIAMETER_CM / 2
CHESTNUT_VOLUME_CM3 = (4/3) * math.pi * (CHESTNUT_RADIUS_CM**3)
CHESTNUT_VOLUME_M3 = CHESTNUT_VOLUME_CM3 * 1e-6

SOLAR_SYSTEM_RADIUS_METER = 4.545e9
SOLAR_SYSTEM_VOLUME_M3 = (4/3) * math.pi * (SOLAR_SYSTEM_RADIUS_METER**3)

volumes = []
minutes = []

def calculate_doubling_times(initial_volume, target_volume):
    doubling_times = 0
    while initial_volume < target_volume:
        initial_volume *= 2
        volumes.append(initial_volume)
        doubling_times += 5
        minutes.append(doubling_times)
    return doubling_times

times_to_fill_solar_system = calculate_doubling_times(CHESTNUT_VOLUME_M3, SOLAR_SYSTEM_VOLUME_M3)

print("It will take {} minutes to cover the solar system with chestnut buns.".format(times_to_fill_solar_system))

plt.figure(figsize=(8, 4))
plt.title("Chestnut Bun Volume Increase Over Time")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")
plt.grid(True)
plt.plot(minutes, volumes)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

BALL_RADIUS_CM = 3.65
BALL_VOLUME_CM3 = (4/3) * math.pi * (BALL_RADIUS_CM**3)
BALL_VOLUME_M3 = BALL_VOLUME_CM3 * 1e-6  # Convert to cubic meters

TOKYO_DOME_VOLUME_M3 = 1240000

volumes = []
minutes = []

def calculate_doubling_times(initial_volume, target_volume):
    doubling_times = 0
    while initial_volume < target_volume:
        initial_volume *= 2
        volumes.append(initial_volume)
        doubling_times += 5
        minutes.append(doubling_times)
    return doubling_times

time_to_fill_tokyo_dome = calculate_doubling_times(BALL_VOLUME_M3, TOKYO_DOME_VOLUME_M3)
print("It will take {} minutes to fill the Tokyo Dome with a ball.".format(time_to_fill_tokyo_dome))

# Plotting
plt.title("Time it takes for Tokyo Dome to fill up")
plt.xlabel("Time (minutes)")
plt.ylabel("Volume (cubic meters)")
plt.tick_params(labelsize=20)
plt.plot(minutes, volumes, color='green', linestyle='dotted', linewidth=5)
plt.show()

```
