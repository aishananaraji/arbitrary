**PROBLEM** **4**

```python
import timeit

def method_1():
    THICKNESS = 0.00008
    Folded_thickness = THICKNESS * 2**43
    print("Thickness: {} meters".format(Folded_thickness))

def method_2():
    THICKNESS = 0.00008
    Folded_thickness = THICKNESS
    for i in range(1, 44):
        Folded_thickness *= 2
    print("Thickness: {} meters".format(Folded_thickness))
```

```python
# Measure execution time using time module
start_time = timeit.default_timer()
method_1()
elapsed_time = timeit.default_timer() - start_time
print("Method 1 execution time: {}[s]".format(elapsed_time))

start_time = timeit.default_timer()
method_2()
elapsed_time = timeit.default_timer() - start_time
print("Method 2 execution time: {}[s]".format(elapsed_time))
```

```python
# Measure execution time using timeit magic command
%%timeit
THICKNESS = 0.00008
Folded_thickness = THICKNESS * 2**43
print("Thickness: {} meters".format(Folded_thickness))
```

```python
# Measure execution time using timeit magic command
%%timeit
Folded_thickness2 = THICKNESS
for i in range(1, 44):
  Folded_thickness2 *=2
print("Thickness: {} meters".format(Folded_thickness2))
```
