###**Problem 1** Creating random numbers

```python
import numpy as np
mean = [-3,0]
cov = [[1.0, 0.8], [0.8, 1]]
np.random.seed(0)
rand_num = np.random.multivariate_normal(mean, cov, size=500)
print(rand_num)
rand_num.shape
```

###**Problem 2** Visualization by scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(rand_num[:, 0], rand_num[:, 1], '.', alpha=0.5 )
plt.title("Visualization by scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis('equal')
plt.show()
```

###**Problem 3** Visualization by histogram

```python
plt.title("Histogram of x1")
plt.xlabel("x1")
plt.ylabel("Frequency")
plt.hist(rand_num[:, 0], bins=30, color='blue', alpha=0.7)
plt.xlim(left=-6, right=2)
plt.show()
```

```python
plt.title("Histogram of x2")
plt.xlabel("x2")
plt.ylabel("Frequency")
plt.hist(rand_num[:, 1], bins=30, color='green', alpha=0.7)
plt.xlim(left=-6, right=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1]]
np.random.seed(0)
rand_num2 = np.random.multivariate_normal(mean, cov, size=500)

plt.title("Two-in-one Scatter Plot")
plt.xlabel("x1")
plt.ylabel("x2")

# Scatter plot for Problem 1
plt.scatter(rand_num[:, 0], rand_num[:, 1], s=20, label='0', marker='o')

# Scatter plot for Problem 4 (corrected)
plt.scatter(rand_num2[:, 0], rand_num2[:, 1], s=20, label='1', marker='o')

plt.legend()
plt.show()
```

###**PROBLEM 5** Data combination

```python
data_combine=np.concatenate((rand_num, rand_num2), axis=0)
print(data_combine)
print(data_combine.shape)
```

```python
combined_data = np.vstack((rand_num, rand_num2))
print(combined_data)
print(combined_data.shape)
```
